```python
from fastai.vision.all import *
```

```python
files = get_image_files('/kaggle/input/forest-fire-dataset')
len(files)
```

```python
def is_fire(f): return f[0] == 'f'
```

```python
dls = ImageDataLoaders.from_name_func('images', files, is_fire, item_tfms=Resize(128))
dls.show_batch()
```

```python
# 50 might be a little excessive
learn = vision_learner(dls, resnet34, metrics=error_rate)
learn.fine_tune(50)
```

```python
learn.show_results()
```
