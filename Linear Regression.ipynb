```python
import pandas as pd
from sklearn.linear_model import LinearRegression
# Define the data
X = pd.DataFrame({'feature1': [1, 2, 3, 4, 5], 'target': [10, 20, 30, 40, 50]})
# Create a linear regression model
lr_model = LinearRegression()
# Fit the model to the data
lr_model.fit(X[['feature1']], X['target'])
# Make predictions on new data
new_data = pd.DataFrame({'feature1': [6, 7, 8, 9, 10]})
predictions = lr_model.predict(new_data[['feature1']])
# Print the predictions
print(predictions)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression


# sample data
study_hours = np.array([2, 3, 4, 6, 7, 8]).reshape(-1, 1)
exam_results = np.array([0, 0, 0, 1, 1, 1])


# Create a Logistic Regression model
logistic_model = LogisticRegression()

# Fit the model to the data
logistic_model.fit(study_hours, exam_results)

# Predict the result for a new study hour value (5)
new_study_hour = np.array([5]).reshape(-1, 1)
predicted_result = logistic_model.predict(new_study_hour)

# Plot the logistic regression curve and data points
x_values = np.linspace(1, 10, 300).reshape(-1, 1)
y_values = 1 / (1 + np.exp(-(logistic_model.intercept_ + logistic_model.coef_[0] * x_values)))

plt.scatter(study_hours, exam_results, color='blue', marker='o', label='Given Data Points')
plt.scatter(new_study_hour, predicted_result, color='red', marker='x', label='Predicted Result for 5 Study Hours')
plt.plot(x_values, y_values, color='green', label='Logistic Regression Curve')
plt.axhline(0.5, color='orange', linestyle='--', label='Decision Boundary (0.5)')
plt.xlabel('Study Hours')
plt.ylabel('Probability of Passing Exam (1)')
plt.title('Logistic Regression Example')
plt.legend()
plt.grid(True)
plt.show()



```
